### Quarta questão

```r
#rgamma(n, shape, rate = 1, scale = 1/rate)

n = 10000
alpha = 1
lambda = 10

gamma = rgamma(n,alpha,lambda)
hist(gamma)

#comparando os valores teoricos e calculados da VAR e EX de gamma
my = mean(gamma)
vargamma = var(gamma)

print("E da gamma:")
print(my)
print("E teórica:")
print(alpha/lambda)

print("Valor esperado da gamma:")
print(vargamma)
print("Valor esperado teórico:")
print(alpha/(lambda^2))

#rexp(n, rate = 1)
exp = rexp(n,lambda)
hist(exp)

var(exp)
var(gamma)
```
